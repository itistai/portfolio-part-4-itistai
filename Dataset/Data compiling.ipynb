```python
df2=df[df.Indicator=='4.1.1']
df2=df2[df2.Sex=='BOTHSEX']
filter=df2.groupby(['GeoAreaName','Type of skill','Education level'])['TimePeriod'].transform('max')
df2=df2[df2.TimePeriod == filter ]
df2=df2[['GeoAreaName','TimePeriod','Value','Type of skill', 'Education level' ]]
filter=df2.groupby(['GeoAreaName','Type of skill'])['Value'].mean().unstack()
df2=pd.DataFrame({'ChildrenMathSkill':filter['SKILL_MATH'],'ChildrenReadingSkill':filter['SKILL_READ']})
pd.options.display.float_format = '{:.1f}%'.format
portfolio4=df2
```

```python
df2=df[df.Indicator=='4.1.2']
df2=df2[df2.Sex=='BOTHSEX']
df2=df2[df2.Quantile=='_T']
df2=df2[df2.Location=='ALLAREA']
filter=df2.groupby(['GeoAreaName','Education level'])['TimePeriod'].transform('max')
df2=df2[df2.TimePeriod == filter ]
df2=df2[['GeoAreaName','TimePeriod','Value','Education level' ]]
filter=df2.groupby(['GeoAreaName'])['Value'].mean()
df2=pd.DataFrame({'SchoolCompletetion':filter})
df2=df2.reset_index()
```

```python
portfolio4=portfolio4.merge(df2,on='GeoAreaName',how='outer')
df2=df[df.Indicator=='4.2.2']
df2=df2[df2.Sex=='BOTHSEX']
filter=df2.groupby(['GeoAreaName'])['TimePeriod'].transform('max')
df2=df2[df2.TimePeriod == filter ]
df2=df2[['GeoAreaName','Value']]
df2 = df2.rename(columns={'Value': 'PreSchoolCompletion'})
portfolio4=portfolio4.merge(df2,on='GeoAreaName',how='outer')
```
